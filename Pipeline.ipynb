```r
def calculate_thickness(ROIs, labelMap, corticalThicknessMap):
    import numpy as np
    import nibabel as nb
    labelMap = nb.load(labelMap)
    labelMapData = labelMap.get_data()
    corticalThicknessMap = nb.load(corticalThicknessMap)
    corticalThicknessData = corticalThicknessMap.get_data()
    results = {}
    for key, value in ROIs.iteritems():
        labelMaskArray = np.zeros(labelMapData.shape)
        label_id = key
        label_name = value
        labelMaskArray[labelMapData == int(label_id)] = 1
        region_average = np.average(corticalThicknessData[ labelMaskArray == 1])
        results[label_name] = region_average
    return results

def save_csv(dictResults, path):
    with open(path, 'w') as csvFile:
        for key, value in dictResults.iteritems():
            csvFile.write('{0},{1:3.10f}\n'.format(key, value))

def calculate_volume(ROIs, labelMap, gmMap):
    import numpy as np
    import nibabel as nb
    labelMap = nb.load(labelMap)
    labelMapData = labelMap.get_data()
    gmMap= nb.load(gmMap)
    gmData= gmMap.get_data()
    results = {}
    
    for key, value in ROIs.iteritems():
        labelMaskArray = np.zeros(labelMapData.shape)
        label_id = key
        label_name = value
        labelMaskArray[labelMapData == int(label_id)] = 1
        #voxel dimension is 1x1x1
        region_volume = np.sum(gmData[ labelMaskArray == 1])
        results[label_name] = region_volume
    return results
```

```r

def calculate_first_volumes(ROIs, path):
    '''
    path to segmented_all_fast_firstseg.nii.gz
    '''
    import numpy as np
    import nibabel as nb
    labelMap = nb.load(path)
    #getting voxel dimensionn
    header = labelMap.get_header()
    voxel_dimension = header.get_zooms()
    voxel_volume = voxel_dimension[0] * voxel_dimension[1] * voxel_dimension[2]
    labelMapData = labelMap.get_data()
    results = {}
    for key, value in ROIs.iteritems():
        labelMaskArray = np.zeros(labelMapData.shape)
        label_id = key
        label_name = value
        labelMaskArray[labelMapData == int(label_id)] = 1
        region_volume = (np.sum(labelMaskArray)) * voxel_volume
        results[label_name] = region_volume
    return results
        
    
```

Permutation function for _p_-value calculation

```r
def permute(X_array, labels_array, n_permutations=1000, random_state=100):
    '''
    fitted_model : a classifier fitted with sklearn
    test_set: array of features that was not used in training the model
    labels: binary labels for test set
    '''
    from sklearn import cross_validation
    import numpy as np
    from sklearn.ensemble import RandomForestClassifier
    rfc = RandomForestClassifier(n_estimators=20)
    X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X_array, labels_array,\
                                                                         test_size=0.5,\
                                                                         random_state=random_state)
    permutation_array = np.zeros((n_permutations, 1))
    for i in xrange(n_permutations):
        if i % 500 == 0:
            print("iteration {0:0d}".format(i))
        permuted_Y_train = np.random.permutation(Y_train)
        permuted_Y_test  = np.random.permutation(Y_test)
        rfc.fit(X_train, permuted_Y_train)
        permuted_accuracy = rfc.score(X_test, permuted_Y_test)
        permutation_array[i] = permuted_accuracy
    return permutation_array

def calculate_p(permutation_array, accuracy):
    '''
    permutation array : numpy array calculated from permute function
    accuracy : mean accuracy from the classifier
    '''
    #numpy documentation website is down so I am not able to access the optimized iterator, I should
    #edit this with C API
    total_length = len(permutation_array)
    counts = 1.
    for element in permutation_array:
        if accuracy <= element:
            counts += 1
    p_value = counts / total_length
    return p_value
```

###Projecting importance array unto brain, there is no tool available to do this, I've written my own here

```r

'''
21  L superior frontal gyrus                                                                                                                                   
22  R superior frontal gyrus
23  L middle frontal gyrus
24  R middle frontal gyrus
25  L inferior frontal gyrus
26  R inferior frontal gyrus
27  L precentral gyrus
28  R precentral gyrus
29  L middle orbitofrontal gyrus
30  R middle orbitofrontal gyrus
31  L lateral orbitofrontal gyrus
32  R lateral orbitofrontal gyrus
33  L gyrus rectus
34  R gyrus rectus
41  L postcentral gyrus
42  R postcentral gyrus
43  L superior parietal gyrus
44  R superior parietal gyrus
45  L supramarginal gyrus
46  R supramarginal gyrus
47  L angular gyrus
48  R angular gyrus
49  L precuneus
50  R precuneus
61  L superior occipital gyrus
62  R superior occipital gyrus
63  L middle occipital gyrus
64  R middle occipital gyrus
65  L inferior occipital gyrus
66  R inferior occipital gyrus
67  L cuneus
68  R cuneus
81  L superior temporal gyrus
82  R superior temporal gyrus
83  L middle temporal gyrus
84  R middle temporal gyrus
85  L inferior temporal gyrus
86  R inferior temporal gyrus
87  L parahippocampal gyrus
88  R parahippocampal gyrus
89  L lingual gyrus
90  R lingual gyrus
91  L fusiform gyrus
92  R fusiform gyrus
101 L insular cortex
102 R insular cortex
121 L cingulate gyrus
122 R cingulate gyrus
165 L hippocampus
166 R hippocampus
'''
ROIs_file = open('/NMOMVPA/template/LPBA40_AIR/maxprob/lpba40_pure_cortex_labels.txt')
ROIs_content = ROIs_file.readlines()
ROIs = {}
for line in ROIs_content:
    line = line.strip()
    ROIs[line.split('\t')[0]] = line.split('\t')[1] 

def back_projector(importance_vector, importance_ROI, atlas_ROI, atlas_file):
    '''
    in:
    
    Gets feature importance array and converts it back to atlas space.
    importance_array : vector of ROI importances
    importance_ROI : vector of random foorest feature imortances
    atlas_ROI: Python dictionary of label IDs along with label names
    
    Out:
    
    Nibabel image in type of nibabel.Nifti1Image
    '''
    import nibabel as nb
    import numpy as np
    atlas = nb.load(atlas_file)
    atlas_affine = atlas.get_affine()
    atlas_data = atlas.get_data()
    importance_map = np.zeros(atlas_data.shape)
    for index, region in enumerate(importance_ROI):
        #atlas roi has space instead of undeline
        atlas_region = region.replace('_', ' ')
        
        
        for key, value in atlas_ROI.iteritems():
            if value == atlas_region:
                label_id = int(key)
                break
        importance_value = importance_vector[index]

        importance_map[atlas_data == label_id] = importance_value
    importance_image = nb.Nifti1Image(importance_map,  atlas_affine)
    return importance_image
```

```r
#importance ROI for back_projector function is the list of column values:
importance_ROI = list(thickness_df.columns.values[0:-1])
atlas_file = '/NMOMVPA/template/LPBA40_AIR/maxprob/pure_cortex.nii.gz'
thickness_importance_image = back_projector(importance_vector=feature_importances_thickness, 
                                  importance_ROI=importance_ROI,
                                  atlas_ROI = ROIs,
                                  atlas_file=atlas_file)
#volume is not included in the manuscript
#volume_importance_image = back_projector(importance_vector=feature_importances_volume, 
#                                  importance_ROI=importance_ROI,
#                                  atlas_ROI = ROIs,
#                                 atlas_file=atlas_file)
surface_importance_image = back_projector(importance_vector=feature_importances_surface, 
                                  importance_ROI=importance_ROI,
                                  atlas_ROI = ROIs,
                                  atlas_file=atlas_file)
```
